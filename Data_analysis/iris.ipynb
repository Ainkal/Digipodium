# handling missing values

```python
#handling missing values
import pandas as pd
import numpy as np 
```

```python
ds = {
    'A':[10,20,np.nan,30,np.nan],
    'B':[1,np.nan,3,np.nan,6],
    'C':[49,56,np.nan,9,54]
}

data=pd.DataFrame(ds)



#checkinng null values in data set
print(data.isnull())
#print the sum of null values per column
print(data.isnull().sum())
```

```python
#removing rows with null values
data.dropna(inplace=True)
data
```

```python
data=pd.DataFrame(ds)
data
```

```python
#removing columns with null vslues
data.dropna(axis=1,inplace=True )
data
```

```python
data
```

```python
dt = {
    'A':[10,20,np.nan,30,np.nan],
    'B':[1,np.nan,3,np.nan,6],
    'C':[49,56,np.nan,9,54]
}

dts=pd.DataFrame(dt)
```

```python
dts.fillna(0,inplace=True)
#replace nan with 0
dts
```

```python
dts=pd.DataFrame(dt)
#dts.copy() 
#fill with column mean
dts['A']=dts['A'].fillna(dts['A'].mean())

dts
```

```python
#fill with column median
dts['B']=dts['B'].fillna(dts['B'].median())
dts
```

```python
#fill with forward fill (previous value)
#dts['C']=dts['C'].fillna(method='ffill')

dts['C'] = dts['C'].ffill()  # âœ… Correct approach

#fill with backward fill (next value)
#dts['B']=dts['B'].fillna(method='bfill')
dts['B'] = dts['B'].bfill()  # âœ… Correct approach
dts
```

 * Apply() function
 

```python
#using apply()

dts['A']=dts['A'].apply(lambda x: x*x)
dts
```
