```python
import matplotlib.pyplot as plt
import pandas as pd


x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o', linestyle='--', color='red', linewidth=2,markersize=8)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Chart")
plt.show()
```

```python
#creating a figure and axes
fig,ax=plt.subplots()
#setting background color
fig.set_facecolor('lightgrey')
ax.set_facecolor('yellow')
x=[1,2,3,4,5,6]
y=[30,20,40,90,10,60]

ax.plot(
    x,y,
    marker='o',
    markersize='2',
    linestyle=':',
    color='black',
    markeredgecolor='blue',
   markerfacecolor='green',
   alpha=0.8 #opacity
)

ax.set_xlabel('x axis',fontsize=14)
ax.set_ylabel('y axis',fontsize=14)
ax.set_title('graph line',fontsize=20)




ax.text(4,90,'point',fontsize=10,color='red')
```
