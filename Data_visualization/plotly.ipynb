```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import plotly.express as px

df = px.data.gapminder()  # Sample dataset
fig = px.scatter(df, x="gdpPercap", y="lifeExp", color="continent", size="pop", log_x=True,template="plotly_dark",)
fig.show()
```

```python
df=pd.read_csv('country_wise_latest.csv')
df.head()
print(df.columns)
```

```python
top_10=df.nlargest(10,'Confirmed')
fig=px.bar(top_10,x='Country/Region',y='Confirmed',color='Confirmed',text='Confirmed',title='Top 10 countries with most confirmed cases',template='plotly_dark',color_continuous_scale=px.colors.sequential.Burg)

fig.show()
```

```python
fig=px.pie(df,names='WHO Region',values='Confirmed',title="Confirmed cases pie chart per WHO region",template='plotly_dark',color_discrete_sequence=px.colors.qualitative.Set2)
fig.show()
```
