```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('country_wise_latest.csv')
df.head()
```

```python


# Load data (replace 'data.csv' with your actual file)


# Select top 10 countries
top_countries = df.nlargest(10, 'Confirmed')

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_countries, hue='Country/Region', palette="Blues_r", legend=False)

plt.title("Top 10 Countries with Highest Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Country")
plt.show()
```

```python
# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])  

# Plot Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of COVID-19 Data")
plt.show()
```
